```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sb
from pylab import rcParams

import scipy
from scipy.stats.stats import pearsonr

%matplotlib inline
rcParams['figure.figsize'] = 5, 4
sb.set_style('whitegrid')

base = pd.read_csv('newDataBaseVk.csv')
base
```

```python
x = base[['countFriends' ,'countFollowers', 'countPhotos', 'countVideos']]
sb.pairplot(x)
```

```python
friends = base['countFriends']
followers = base['countFollowers']
photos = base['countPhotos']
videos = base['countVideos']

print('Рассчет корреляции Пирсона')

pearsonr_coefficient, p_value = pearsonr(friends, followers)
print( 'Коэфицент корреляции Пирсона количество друзей/количество подписчиков %0.3f' % (pearsonr_coefficient))

pearsonr_coefficient, p_value = pearsonr(friends, photos)
print( 'Коэфицент корреляции Пирсона количество друзей/количество фотографий %0.3f' % (pearsonr_coefficient))

pearsonr_coefficient, p_value = pearsonr(friends, videos)
print( 'Коэфицент корреляции Пирсона количество друзей/количество видео %0.3f' % (pearsonr_coefficient))
```
